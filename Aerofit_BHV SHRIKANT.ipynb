#CUSTOMER PROFILING

```python
#Age Profiling

def age_profile(age):
  if (age >= 18) & (age < 25):
    return '18 - 25'
  elif (age >= 25) & (age < 30):
    return '25 - 30'
  elif (age >= 30) & (age < 35):
    return '30 - 35'
  elif (age >= 35) & (age < 45):
    return '35 - 45'
  else:
    return '45 and above'

df['Age_profile'] = df['Age'].apply(age_profile)
df
```

```python
# Income Profiling


def income_profile(income):
  if (income >= 29000) & (income < 45000):
    return 'Lower Income'
  elif (income >= 45000) & (income < 65000):
    return 'Middle Income'
  elif (income >= 65000):
    return 'Higher Income'


df['Income_profile'] = df['Income'].apply(income_profile)
df
```

```python
# Education Profiling


def education_profile(education):
  if (education >= 12) & (education < 14):
    return 'Basic Education'
  elif (education >= 14) & (education <= 16):
    return 'Intermediate Education'
  elif (education > 16):
    return 'Advanced Education'


df['Education_profile'] = df['Education'].apply(education_profile)
df
```

```python
sns.stripplot(data=df, x = 'Income_profile', y = 'Age_profile', hue = 'Product')
```

OBSERVATION :

*****************************

1.   Customers at Lower Income levels and belonging to age group 18-25 choose KP481 over other models.

2.   Customers at Lower Income levels and belonging to age group 25-30 choose KP281 over other models.

3.   Customers at Middle Income levels and belonging to age group 18-25 choose KP781 over other models.

4.   Customers at Middle Income levels and belonging to age group 25-30 generally don't choose KP281.

5.   Customers at Middle Income levels and belonging to age group 30-45 choose KP481 over other models.

6.   Customers at Higher Income levels and belonging to any age group choose KP781 over other models.

# RECOMMENDATIONS


1.   KP281 is the top-selling product with 44.44% contribution in total sales. So, they should try pushing this entry level product at emerging markets. Also, they should run trial offers so that new customers can get into the world of AEROFIT products.

2.   KP781 is bought exclusively by customers of age group 25 to 30. So, we can do target advertisment accordingly.

3.   KP781 is also the most bought product among Advanced Education profile or Higher Income profile or both. So, we can focus on these details first while trying to understand customer requirements and recommending a product.

4.   Among people with low to moderate fitness levels, KP281 & KP481 are popular. Whereas, KP781 is the preferred choice for fitness enthusiasts. So, we can offer few other low-priced fitness accessories with KP781.

5.   Customers with high Usage prefers KP781 so we can bundle certain maintenance packs along with the product.
